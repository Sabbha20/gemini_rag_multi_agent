## LangChain Agents
https://docs.langchain.com/oss/python/langchain/middleware/built-in



```python
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.agents import create_agent
from langchain.messages import SystemMessage, HumanMessage
from zmaza_scripts import base_tools, llm_models
```

```python
model = ChatGoogleGenerativeAI(model=llm_models.gemini2)
```

```python
system_prompt = "You are a financial analyst specializing in tech stocks.\nProvide data-driven analysis with clear insights. you have access to web_search tools and get_weather tools."
```

### Basic Agent

```python
agent = create_agent(
    model=model,
    tools=[base_tools.z_web_search, base_tools.z_get_weater],
    system_prompt=system_prompt
)
agent
```
