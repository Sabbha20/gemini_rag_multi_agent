## Multimodal 

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.messages import AIMessage, HumanMessage, SystemMessage
```

```python
gemini3 = "gemini-3-pro-preview"
gemini2 = "gemini-2.5-flash"

system_msg = SystemMessage(content="You are a helpful assistant.")
human_msg = HumanMessage(
    [
        {
            "type" : "text",
            "text": "Describe the image in detail."
        },
        {
            "type": "image",
            "url": "https://static.vecteezy.com/system/resources/thumbnails/051/599/550/small/cute-baby-hippopotamus-submerged-in-water-surrounded-by-lush-green-foliage-concept-of-wildlife-nature-africa-photo.jpg"
        }
        
    ]
)
```

```python
model = ChatGoogleGenerativeAI(model=gemini3)
response = model.invoke([system_msg, human_msg])
```

```python
response
```

```python
print(response.text)
```

## Image from local file

```python
import base64

mime_type = "image/png"

img_bytes = open("data/hippo.png", "rb").read()
base64_bytes = base64.b64encode(img_bytes).decode("utf-8")

base64_bytes
```

```python
model = ChatGoogleGenerativeAI(model=gemini2)

human_msg = HumanMessage(
    [
        {
            "type" : "text",
            "text": "Describe the image in detail."
        },
        {
            "type": "image",
            "base64": base64_bytes,
            "mime_type": mime_type
        }
        
    ]
)

response = model.invoke([system_msg, human_msg])
```

```python
print(response.text)
```
